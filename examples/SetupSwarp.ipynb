# OK now create all of the commands of interest

Generally speaking one expects that the commands will be the same for all tiles of a given field


# This is an attempt to systemetize the creation of all the commands to run a single tile



```python
def c42_commands(tile=3,default=default):
    field='LMC_c42'
    summarize(field)
    create_swarp_command(field=field,tile=tile,filt='Ha',exp=[30,800],defaults=default)
    create_swarp_command(field=field,tile=tile,filt='Ha',exp=[800],defaults=default)
    create_swarp_command(field=field,tile=tile,filt='SII',exp=[60,800],defaults=default)
    create_swarp_command(field=field,tile=tile,filt='SII',exp=[800],defaults=default)
    create_swarp_command(field=field,tile=tile,filt='R',exp=[8,60],defaults=default)
    create_swarp_command(field=field,tile=tile,filt='R',exp=[60],defaults=default)
    create_swarp_command(field=field,tile=tile,filt='N708',exp=[60,400],defaults=default)
    create_swarp_command(field=field,tile=tile,filt='N708',exp=[400],defaults=default)
    return

c42_commands()
    
```

```python
def c45_commands(tile=1):
    field='LMC_c45'
    summarize(field)
    create_swarp_command(field=field,tile=tile,filt='Ha',exp=[30,800],defaults=default)
    create_swarp_command(field=field,tile=tile,filt='Ha',exp=[800],defaults=default) 
    
    create_swarp_command(field=field,tile=tile,filt='SII',exp=[60,400],defaults=default)
    create_swarp_command(field=field,tile=tile,filt='SII',exp=[400],defaults=default) 
    
    create_swarp_command(field=field,tile=tile,filt='R',exp=[8,30],defaults=default)
    create_swarp_command(field=field,tile=tile,filt='R',exp=[30],defaults=default) 
    
    create_swarp_command(field=field,tile=tile,filt='N708',exp=[30,400],defaults=default)
    create_swarp_command(field=field,tile=tile,filt='N708',exp=[400],defaults=default) 
    
    


c45_commands()
    
```
